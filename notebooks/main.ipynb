```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/dirty_cafe_sales.csv')
```

```python
df.sample(15)
```

## Obervações
- Nomes das colunas não estão em snake_case
- Campos em quantidade como 'UNKNOWN'
- Campos em 'price per unit como 'UNKNOWN'
- Alguns campos em 'Location' estão como Nan e outros como ERROR
- Todas as colunas são do tipo 'object' 

```python
print(df.info())
print(df.columns)
```

## Normalizando as colunas

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
df
```

# Substituindo valores faltantes por NaN
- O dataset usa valores como 'ERROR' para se referir a valores que estão faltando. Isso não permite que possamos fazer cálculos futuros na coluna em questão, diferente de 'NaN', onde o podemos tratá-lo de maneira nativa

```python
 
valores_invalidos = ['ERROR','UNKNOWN','nan', '']

df = df.replace(valores_invalidos,np.nan)
df.sample(10)
```

## Transformando os tipos das colunas
- Como foi percebido na etapda de análise, todas as colunas estão como tipo 'object'. Temos que transformá-las nos tipos apropriados

```python
#Transformando colunas para float
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors='coerce')
df['total_spent'] = pd.to_numeric(df['total_spent'], errors='coerce')

# Transformação para número inteiro
df['quantity'] = pd.to_numeric(df['quantity'],errors ='coerce').astype('Int64')

# Transformação para data
df['transaction_date'] = pd.to_datetime(df['transaction_date'],errors ='coerce')

print(df.dtypes)
df.sample(10)
```

# Imputação Lógica
- Alguns campos estão faltantes, mas podem ser preenchidos, pois temos informações suficientes para realizar essa tarefe
- Tome como exemplo o item 'Cake', onde, em certos momentos, temos o seu preço (3.0), mas em outros casos não temos. Como o preço do café não é desconhecido, podemos resolver esse problema de dados faltantes

```python
precos = {
    'Coffee': 2.0, 'Tea': 1.5, 'Juice': 3.0, 'Smoothie': 4.0,
    'Sandwich': 4.0, 'Cake': 3.0, 'Cookie': 1.0, 'Salad': 5.0
}

df['price_per_unit'] = df['price_per_unit'].fillna(df['item'].map(precos))

df['total_spent'] = df['quantity'] * df['price_per_unit']
#df['quantity'] = df['total_spent'] / df['price_per_unit']
df.dropna(subset=['quantity','total_spent'],how='all',inplace=True)
df.dropna(subset=['item'],inplace=True)
df.sample(10) 
```

```python
# Preechendo valores categóricos nulos
df['payment_method'] = df['payment_method'].fillna('unknown')
df['location'] = df['location'].fillna('unknown')
df.sample(10)
```

## Checagem final


```python
duplicatas = df.duplicated(subset=['transaction_id']).sum()
duplicatas
if duplicatas > 0:
    df.drop_duplicates(subset=['transaction_id'],keep='first',inplace=True)
```
